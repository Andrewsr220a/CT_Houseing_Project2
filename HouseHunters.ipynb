```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

from sqlalchemy import create_engine
```

```python
#Loading csv files

state_prices_file = "Sale_Prices_State.csv"
city_prices_file = "Sale_Prices_City.csv"
city_prices_2_file = "Sale_Prices_City 2.csv"
```

```python
state_prices_df = pd.read_csv(state_prices_file)
state_prices_df.head()
```

```python
#State 2 data from 2018 - 2020

state_prices_2018_df  = state_prices_df.loc[:, ['RegionID', 'RegionName', 'SizeRank',"2018-01",'2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12','2020-01','2020-02','2020-03']]

state_prices_2018_df.rename(columns = {'RegionName':'StateName'}, inplace = True)
state_prices_2018_df
```

```python
# add new column with average sale for each state
state_prices_2018_df['Average Sale'] = state_prices_2018_df.mean(axis=1)
# display the dataframe
state_prices_2018_df
```

```python
#State 2 data for 2019

state_prices_2019_df  = state_prices_df.loc[:, ['RegionID', 'RegionName', 'SizeRank','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12']]
state_prices_2019_df.rename(columns = {'RegionName':'StateName'}, inplace = True)
state_prices_2019_df
```

```python
# add new column with average sale for each state
#state_prices_2019_df['Average Sale'] = state_prices_2019_df.mean(axis=1)
state_prices_2019_df['Average Sale']  = state_prices_2019_df[['2019-01', '2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12']].mean(axis=1)
# display the dataframe
state_prices_2019_df
```

```python
city_prices_df = pd.read_csv(city_prices_file)
city_prices_df.head()
```

```python
#City data from 2018- 2020 df

city_prices_2018_df = city_prices_df.loc[:, ['RegionID', 'RegionName', 'StateName', 'SizeRank',"2018-01",'2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12','2020-01','2020-02','2020-03']]
city_prices_2018_df
```

```python
city_prices_2_df = pd.read_csv(city_prices_2_file)
city_prices_2_df.head()
```

```python
#City 2 data from 2018- 2020 df

city_prices_2_2018_df = city_prices_2_df.loc[:, ['RegionID', 'RegionName', 'StateName', 'SizeRank',"2018-01",'2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08','2018-09','2018-10','2018-11','2018-12','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12','2020-01','2020-02','2020-03']]
city_prices_2_2018_df
```

```python
# add new column with average sale for each region of the state
city_prices_2_2018_df['Average City Sale'] = city_prices_2_2018_df.mean(axis=1)
# display the dataframe
city_prices_2_2018_df
```

```python
#City 2 data from 2019 df

city_prices_2_2019_df = city_prices_2_df.loc[:, ['RegionID', 'RegionName', 'StateName', 'SizeRank','2019-01','2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12']]
city_prices_2_2019_df.rename(columns = {'StateName':'StateName'}, inplace = True)
city_prices_2_2019_df
```

```python
# add new column with average sale for each region of the state
city_prices_2_2019_df['Average City Sale']  = city_prices_2_df[['2019-01', '2019-02','2019-03','2019-04','2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12']].mean(axis=1)
# display the dataframe
city_prices_2_2019_df
```

Create Database Connection

```python
connection_string = "postgres:postgres@localhost:5432/StateCityHousingPrice_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Checking the tables 
engine.table_names()
```

```python
# Use pandas to load csv converted DataFrame into database

state_prices_2019_df.to_sql(name='statesales', con=engine, if_exists='append', index=False)
```

```python
#Use pandas to load csv converted DataFrame into database

city_prices_2_2019_df.to_sql(name='citysales', con=engine, if_exists='append', index=False)
```

```python
# ### Confirm data has been added by querying the statesales table

pd.read_sql_query('select * from statesales', con=engine).head()
```

```python
### Confirm data has been added by querying the citysales table

pd.read_sql_query('select * from citysales', con=engine).head()
```
